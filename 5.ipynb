```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('train.csv', index_col='id', parse_dates = ['pickup_datetime'])
test = pd.read_csv('test.csv', index_col='id', parse_dates = ['pickup_datetime'])
from sklearn.preprocessing import MinMaxScaler
mmsc = MinMaxScaler()
```

```python
from catboost import CatBoostRegressor
```

```python
df = df1.drop("trip_duration", axis = 1)
```

```python
columns = ["passenger_count", "store_and_fwd_flag", "vendor_id"]
```

```python
df.head()
```

```python
special = pd.DataFrame(mmsc.fit_transform(df[['pickup_longitude']]), 
                    index = df.index,
                    columns = ['pickup_longitude'])
df = pd.concat([df.drop('pickup_longitude', axis=1), special], axis=1)
```

```python
special2 = pd.DataFrame(mmsc.fit_transform(df[['pickup_latitude']]), 
                    index = df.index,
                    columns = ['pickup_latitude'])
df = pd.concat([df.drop('pickup_latitude', axis=1), special2], axis=1)
```

```python
special3 = pd.DataFrame(mmsc.fit_transform(df[['dropoff_longitude']]), 
                    index = df.index,
                    columns = ['dropoff_longitude'])
df = pd.concat([df.drop('dropoff_longitude', axis=1), special3], axis=1)
```

```python
special4 = pd.DataFrame(mmsc.fit_transform(df[['dropoff_latitude']]), 
                    index = df.index,
                    columns = ['dropoff_latitude'])
df = pd.concat([df.drop('dropoff_latitude', axis=1), special4], axis=1)
```

```python
trip_duration = df1["trip_duration"]
```

```python
CatBoost = CatBoostRegressor(cat_features = columns) #depth = 11
```

```python
df = df.fillna(0)
```

```python
Cat = CatBoost.fit(df, trip_duration)
```
